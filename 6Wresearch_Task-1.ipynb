## 5. find the 95%, 90%, and 99%, confidence interval for 'Potential','wage','weight' column.

```python
import scipy.stats as stats

# Specify the columns for which we want to calculate confidence intervals
columns_to_analyze = ['Potential', 'Wage', 'Weight']
# df['Release Clause'] = df['Release Clause'].str.replace('€', '').str.replace('M', '').astype(float) * 1000000
if(isinstance(df['Weight'][0],str)):
    df['Weight'] = df['Weight'].str.replace('lbs', '').astype(float)* 0.453592
# Set the desired confidence levels
confidence_levels = [0.90, 0.95, 0.99]

# Function to calculate confidence intervals
def calculate_confidence_interval(data, confidence_level):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use Bessel's correction for sample standard deviation
    n = len(data)
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(n))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Calculate and print the confidence intervals for each specified column
for column in columns_to_analyze:
    if column in df.columns:
        for confidence_level in confidence_levels:
            lower_bound, upper_bound = calculate_confidence_interval(df[column], confidence_level)
            print(f"{column} {int(confidence_level * 100)}% Confidence Interval:")
            print(f"Lower Bound: {lower_bound:.2f}")
            print(f"Upper Bound: {upper_bound:.2f}")
            print()
    else:
        print(f"Column '{column}' does not exist in the DataFrame.")

```

## 6.Prove Central Limit Theorom by using 'potential' column of the game_data.
### The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that, given certain conditions, the arithmetic mean of a sufficiently large number of iterates of independent random variables, each with a well-defined expected value and well-defined variance, will be approximately normally distributed, regardless of the underlying distribution.

In data science, the CLT is used to make inferences about the population mean based on a sample mean. For example, if you have a sample of data from a population, you can use the CLT to calculate the probability that the sample mean is within a certain range of the population mean.

```python


# Extract the 'Potential' column as a NumPy array
potential_data = df['Potential'].values

# Number of samples to take and sample size
num_samples = 1000
sample_size = 30

# Function to calculate the sample mean for a given sample
def calculate_sample_mean(sample):
    return np.mean(sample)

# Create an empty list to store the sample means
sample_means = []

# Take random samples and calculate the sample means
for _ in range(num_samples):
    sample = np.random.choice(potential_data, size=sample_size, replace=False)
    sample_mean = calculate_sample_mean(sample)
    sample_means.append(sample_mean)

# Plot the histogram of sample means using Plotly
fig = px.histogram(sample_means, nbins=30, labels={'value': 'Sample Mean of Potential'},
                   title='Distribution of Sample Means (Central Limit Theorem)',)
fig.show()

# Create a line plot of the sample means using Plotly
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(num_samples), y=sample_means, mode='lines', line=dict(color='lightgreen')))
fig.update_layout(title='Distribution of Sample Means (Central Limit Theorem)',
                  xaxis_title='Sample Number', yaxis_title='Sample Mean of Potential')
fig.show()
```
